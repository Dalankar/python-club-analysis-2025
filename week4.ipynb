```python
plt.figure(figsize=(5, 7))
sns.histplot(data=df, x='age')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age (Years)')
plt.ylabel('Number of patients')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='days_to_recovery')
plt.title('Summary of patient recovery time')
plt.xlabel('Days to recovery')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='c_reactive_protein', y='days_to_recovery', hue='bmi')
plt.title('Inflammation vs Recovery time by treatment group')
plt.xlabel('C-reactive protein')
plt.ylabel('Days to recovery')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='sex', y='days_to_recovery', hue='sex')
plt.title('Comparison of Recovery time by treatment group')
plt.xlabel('Treatment group')
plt.ylabel('Days to full recovery')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='PuRd', linewidths=0.5)
plt.title('Correlation heatmap of patient data')
plt.show()
```
